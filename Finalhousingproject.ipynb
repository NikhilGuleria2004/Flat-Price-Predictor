```python

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


df = data


df['total rooms'] = df['bedrooms'] + df['bathrooms'] + df['stories']


columns_to_drop = ['bedrooms', 'bathrooms', 'stories']
df = df.drop(columns=columns_to_drop)


skewed_features = ['price', 'area']
df[skewed_features] = np.log1p(df[skewed_features])  

cleaned_data_filename = 'cleaned_house.csv'
df.to_csv(cleaned_data_filename, index=False)
print(f"Preprocessed and cleaned data saved as {cleaned_data_filename}")


X = df.drop('price', axis=1)
y = df['price']

random_state = 73
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)


cat_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), cat_columns)
    ],
    remainder='passthrough'
)


pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])


pipeline.fit(X_train, y_train)


y_pred = pipeline.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")


r2 = r2_score(y_test, y_pred)


r2_percentage = r2 * 100

print(f"R-squared Score: {r2_percentage:.2f}%")
```

```python
dt  = pd.read_csv('Housing.csv')
corr_order = dt.corr().loc['area':, 'price'].abs().sort_values(ascending=False)
top_vars = list(corr_order[:10].index)
top_vars
X_data = dt.loc[:, top_vars]
y_data = dt.loc[:, 'price']

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, shuffle=True, random_state=12)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
sc = MinMaxScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

print('intercept: ', np.round(lr.intercept_, 4))
for i, val in enumerate(lr.coef_):
    print(f'coef for {df.columns[i+1]}: {np.round(val, 4)}')
```

```python
y_pred = lr.predict(X_test)
residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)  # Add a horizontal line at y=0
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
X_test = pd.DataFrame(data=X_test, columns=top_vars)
```

```python
models = [
    LinearRegression(),
    DecisionTreeRegressor(max_depth=3, random_state=11),
    RandomForestRegressor(max_depth=3, random_state=11)]
for model in models:
    model_name = type(model).__name__

    if 'LinearRegression' in str(type(model)) or 'DecisionTreeRegressor' in str(type(model)):
        pipeline = make_pipeline(MinMaxScaler(), model)
    else:
        pipeline = model

    pipeline.fit(X_train, y_train)

    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    
    print(f'{model_name} - Train MSE: {train_mse:.5f} | Test MSE: {test_mse:.5f}')
    print('-' * 65)
    
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=y_test, y=y_test, color='blue', label='Actual Values')
    sns.scatterplot(x=y_test, y=y_test_pred, color='orange', label='Predicted Values')
    plt.title(f'Predicted vs. Actual')
    plt.legend()
    plt.show()
```
